```python
model = genai.GenerativeModel('gemini-1.0-pro-latest')
```

```python
safety_settings = [
    {
        "category": "HARM_CATEGORY_DANGEROUS",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_HARASSMENT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_HATE_SPEECH",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
        "threshold": "BLOCK_NONE",
    },
    {
        "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
        "threshold": "BLOCK_NONE",
    },
]
```

SAMPLE VEDIKETTU

```python
script = "You are an expert letter writer tasked with generating formal request letters for various situations. Each letter should be unique, well-structured, and tailored to the specific scenario provided. The letter should be around 100-120 words in length and maintain a professional and polite tone."

prompts = ["Write a request letter to a university admissions office seeking an extension for submitting application materials due to extenuating circumstances.", "Draft a letter requesting a medical leave of absence from an employer due to a significant health issue that requires an extended period of recovery.", "Compose a letter to a college dean requesting an educational leave of absence for one semester to pursue a professional opportunity related to your field of study.", "Create a letter to a school principal requesting permission for your child to be excused from classes for a family vacation during the upcoming school term.", "Write a request letter to a university housing office seeking an accommodation change due to a personal or medical issue that necessitates a different living arrangement."]
```

```python
for i in prompts:    
    response = model.generate_content(f'{script} {i}', safety_settings=safety_settings)
    print(response.text)
```

PROMPTING

```python
import csv
import time

with open('./prompts.txt', 'r') as f:
    prompts = f.readlines()

with open('prompts_responses.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['prompt', 'response']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for prompt in prompts:
        prompt_parts = prompt.split('. ', 1)
        prompt_number = prompt_parts[0]
        prompt_text = prompt_parts[1].strip()

        print(f"{prompt_text}")
        response = model.generate_content(f'{script} {prompt_text}', safety_settings=safety_settings)
        print(response.text)

        writer.writerow({'prompt': prompt_text, 'response': response.text})
        time.sleep(5)
```
